```python
import matplotlib.pyplot as plt
```

Write a python script to display a bar chart for the top data analyst skills using the rankings dataset below
Data:
- Power BI=80
- SQL=60
- Python=40
- Excel=40
- Statistics=20

```python
ranks=[80,60,40,40,30]
skills=["Python","SQL","Power BI","Excel","Statistics"]
c=["red","blue","yellow","black","green"]
plt.bar(skills,ranks,color=c)
plt.title("Top Data Analyst Skills")
plt.xlabel("Skills",color="blue")
plt.ylabel("Ranks")
plt.show()
```

```python
x=["science","commerce","arts"]
y=[200,300,500]
c=["red","blue","green"]
plt.bar(x,y,color=c,linewidth=5,edgecolor="black")
plt.xlabel("courses")
plt.ylabel("students enrolled")
plt.title("students enrolled for different courses 2020")
plt.show()
```

## Line graph

```python
x=[10,20,30,40,50]
y=[10,35,60,75,90]
z=[25,40,65,80,105]
plt.plot(x,y,color="red")
plt.plot(x,z,color="green")
plt.title("Test Plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["is y","is z"])
plt.show()
```

## Simple Bar Graph

```python

x=["english","math","kiswahili","biology","chemistry"]
y=[65,95,78,80,75]
c=["black","blue","green","yellow","red"]
plt.bar(x,y,width=0.8,color=c)
plt.xlabel("subjects",color="blue")
plt.ylabel("mean score",color="blue")
plt.title("students mean score in different subjects")
plt.show()
```
